Heatmap

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(low=1,high=100,size=(10,10))
data
```

```python
sns.heatmap(data,vmin=30,vmax=70)
plt.show()
```

```python
sns.heatmap(data,cmap="tab20")
plt.show()
```

```python
sns.heatmap(data,cmap="tab20",annot=True)
plt.show()
```

```python
sns.heatmap(data,cmap="Pastel1",vmin=10,vmax=80,annot=True,linewidths=1,linecolor='black',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
df = pd.DataFrame(np.random.random((5,5)),columns=["A","B","C","D","E"])
sns.heatmap(df,annot=True)
plt.show()
```

```python
data = np.random.randint(low=1,high=100,size=(3,10))
data
```

```python
plt.Figure(figsize=(20,3))
sns.heatmap(data,cmap="Pastel1",vmin=10,vmax=80,annot=True,linewidths=1,linecolor='black',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

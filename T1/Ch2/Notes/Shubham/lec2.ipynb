```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
import plotly.express as px
import plotly.io as pio
```

```python
df = pd.read_csv("../../../Ch1/datasets/auto-mpg.csv")
df
```

```python
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
```

```python
df = df.dropna(subset=["horsepower"])
plt.figure(figsize=(10, 6))
sns.scatterplot(x="horsepower", y="mpg", hue="origin",style="origin",size="weight", data=df)
plt.title("Horsepower vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("Miles Per Gallon (MPG)")
plt.legend(title="Origin")
plt.show()
```

NetworkX

directed graph : it has arrow direction

undirected graph : it does not arrow direction

```python
G = nx.Graph()

G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))

G.add_edge(1, 2)
G.add_edge(1,1)
G.add_edges_from([(2,3),(3,6),(4,6),(5,6)])

nx.draw_networkx(G)
plt.show()
```

```python
g= nx.DiGraph()
g.add_nodes_from([1,2,3,4,5])
g.add_edges_from([(1,2),(1,3),(2,4),(3,4),(4,5)])
nx.draw(g, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, arrowsize=20)
plt.show()
```

```python
g =nx.DiGraph()

# g.add_nodes_from(range(1,4))
l = [[0,1,0,1], [0,0,0,1], [1,0,0,0], [1,0,1,0]]

for i in range(4):
    for j in range(4):
        if l[i][j] == 1:
            g.add_edge(i, j)

nx.draw(g, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, arrowsize=20)
plt.show()
```

```python
pio.renderers.default = "browser"

data = {
    "Department": ["IT", "IT","HR","HR", "Finance","Finance"],
    "Gender": ["M","F","M","F","M","F"],
    "Salary": [70000, 75000, 50000, 55000, 80000, 85000],
    "role": ["Manager", "Analyst", "Manager", "Analyst", "Manager", "Analyst"]
}

df = pd.DataFrame(data)
px.bar(df, x="Department", y="Salary", color="Gender", hover_data=["role"],barmode="stack", title="Salary Distribution by Department and Gender").show()
```

```python
# 2dscatter plot with plotly

px.scatter(df, x="Department", y="Salary", color="Gender", facet_row="Department", facet_col="Gender", title="Salary Distribution by Department and Gender").show()
```

```python
# 3d scatter plot with plotly
px.scatter_3d(df, x="Department", y="Salary", z="role", color="Gender", title="3D Scatter Plot of Salary by Department and Role").show()
```

```python
# scatter_matrix with plotly
px.scatter_matrix(df, dimensions=["Department", "Salary", "role"],  color="Gender", title="Scatter Matrix of Salary by Department and Role").show()
```

```python
# line plot with plotly
px.line(df, x="Department", y="Salary", color="Gender",markers=True, title="Line Plot of Salary by Department and Gender").show()
```

```python
# histogram with plotly

px.histogram(df, x="Salary", color="Gender",nbins=1, title="Salary Distribution by Gender").show()
```

```python
# pie chart with plotly

px.pie(df, names="Department", values="Salary", title="Salary Distribution by Department").show()  
```

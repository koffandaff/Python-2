### How to Find Outliers

Interquartile range (IQR) = Q3(75%) - Q1(25%) 

Now using eq we define a range a lower and a hgiher range if the value is out of that range then it's an outlier

lower_value => q1 - 1.5 * IQR 
higer_value => q3 + 1.5 * IQR

```python
def find_outliers(ds,col):
    q1 = ds[col].quantile(0.25) # fidnging 25%
    q3 = ds[col].quantile(0.75)
    IQR = q3 - q1
    low_val = q1 - 1.5 * IQR
    upper_val = q3 + 1.5 * IQR
    print(f'Low: {low_val}, High: {upper_val}')

    ds = ds.loc[(ds[col] < low_val) | (ds[col]> upper_val)]
    return ds

df = pd.read_csv('../../datasets/auto-mpg.csv')
find_outliers(df,'mpg')
```

```python
find_outliers(df,'acceleration')
```

```python
import numpy as np
import pandas as pd

def remove_outliers(ds,col):
    q1 = ds[col].quantile(0.25) # fidnging 25%
    q3 = ds[col].quantile(0.75)
    IQR = q3 - q1
    low_val = q1 - 1.5 * IQR
    upper_val = q3 + 1.5 * IQR
    print(f'Low: {low_val}, High: {upper_val}')

    ds = ds.loc[(ds[col] > low_val) & (ds[col] < upper_val)]
    return ds

df = pd.read_csv('../../datasets/auto-mpg.csv')
remove_outliers(df,'mpg')
```

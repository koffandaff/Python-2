## Data Cleaning

1. missing values
2. methods to remove missing values:
    1. Delete complete row with NaN
    2. fill with dummy data
    3. replace it with mean, median, mode

```python
sales_data = pd.DataFrame({'Name': ['William', 'Emma', 'Sofia', 'Markus', 'Edward', 'Thomas', 'Ethan', np.nan , 'Arun', 'Ankita', 'Paule'],
                            'Region': [np.nan, 'North', 'East', np.nan, 'West', 'West', 'South', np.nan, 'West', 'East', 'South'],
                            'Sales': [np.nan, 12000, 15000, np.nan, 18000, 20000, 22000, np.nan, 25000, 28000, 30000],
                            'Expense': [np.nan, 1000, 1500, np.nan, 2000, 2500, 3000, np.nan, 3500, 4000, 4500]})

sales_data
```

```python
sales_data.isna() # isnull() is also valid
```

```python
sales_data.isna().sum()
```

```python
sales_data.dropna()
```

```python
sales_data.dropna(thresh=2) # thresh means to check how many valid data is present
```

```python
sales_data.dropna(how='all')
```

```python
sales_data.dropna(subset=['Sales','Expense'])
```

```python
sales_data.dropna(axis=0) #normal check null value
```

```python
sales_data.dropna(axis=1) #normal check null value
```

```python
sales_data
```

```python
sales_data.fillna(0)
```

```python
sales_data['Name'].fillna('Guest')
```

```python
sales_data['Sales'].fillna(sales_data['Sales'].mean())
```

```python
sales_data['Sales'].fillna(sales_data['Sales'].median())
```

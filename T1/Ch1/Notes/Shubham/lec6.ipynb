```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

concat function

```python
df1 = pd.DataFrame({"A":[1,2],"B":[3,4]})
df2 = pd.DataFrame({"A":[5,6],"C":[7,8]})
result = pd.concat([df1,df2],axis=0)
result
```

```python
result = pd.concat([df1,df2],axis=1)
result
```

```python
result = pd.concat([df1,df2],axis=0,join="outer")
result
```

```python
result = pd.concat([df1,df2],axis=0,join="inner")
result
```

```python
result = pd.concat([df1,df2],axis=0,ignore_index=True)
result
```

```python
df1 = pd.DataFrame({"E":[1,2],"B":[3,4]})
df2 = pd.DataFrame({"C":[5,6],"D":[7,8]})
result = pd.concat([df1,df2],sort=True)
result
```

merge function

```python
df1 = pd.DataFrame({'Employee': ['A','B','C'],
                    'Department': ['IT', 'HR', 'Finance']})

df2 = pd.DataFrame({'Employee': ['A','B','D'],
                    'Salary': [50000,4000,30000]})

result = pd.merge(df1,df2,on ="Employee")
result
```

```python
result = pd.merge(df1,df2,on ="Employee",how="outer")
result
```

```python
result = pd.merge(df1,df2,on ="Employee",how="left")
result
```

```python
result = pd.merge(df1,df2,on ="Employee",how="right")
result
```

```python
dt1 = pd.DataFrame({
    'Emp_ID': [101,102,103],
    'Name': ['Amit', 'Riya', 'John'],
    'Salary': [50000, 66000, 55000]
})
dt2 = pd.DataFrame({
    'Emp_ID': [101,102,104],
    'Name': ['Amit', 'Riya', 'David'],
    'Salary': [57000, 96000, 75000]
})

result = pd.merge(dt1, dt2, on='Emp_ID')
result
```

```python

result = pd.merge(dt1, dt2, on='Emp_ID',suffixes=('_left', '_right'))
result
```

```python
dt1 = pd.DataFrame({
    'Emp_ID': [101,102,103],
    'Name': ['Amit', 'Riya', 'John']
})
dt2 = pd.DataFrame({
    'ID': [101,102,104],
    'Salary': [57000, 96000, 75000]
})

result = pd.merge(dt1, dt2, left_on='Emp_ID', right_on='ID')
result
```

```python
dt1 = pd.DataFrame({
    'Name': ['Amit', 'Riya', 'John'],
     "index":[100,101,102]
     })
dt2 = pd.DataFrame({
    'Salary': [57000, 96000, 75000]
},index=[100,101,102])

result= pd.merge(dt1,dt2, left_on="index", right_index=True)
result
```

groupby()

```python
df = pd.DataFrame({
    'Product Category': ["Elec","Furni","Elec","Furni"],
    'Sales': [200,150,300,100]
})
df
```

```python
result = df.groupby('Product Category')["Sales"].sum()
result
```

```python
result = df.groupby('Product Category')["Sales"].mean()
result
```

```python
result = df.groupby('Product Category')["Sales"].median()
result
```

```python
result = df.groupby('Product Category')["Sales"].describe()
result
```

```python
result = df.groupby('Product Category')["Sales"].sample(2)
result
```

```python
result = df.groupby('Product Category')["Sales"].groups
result
```

```python
result = df.groupby('Product Category',as_index=True,sort=True)["Sales"].count()
result
```

```python
result = df.groupby('Product Category',as_index=False,sort=True)["Sales"].count()
result
```

```python
df = pd.DataFrame({
    "region": ["East","East","West","West","East","West"],
    "Sales": [200,150,300,100,250,350],
    "Manager":["A","B","C","D","E","F"]
})
df
```

```python
def sales_range(x):
    return x['Sales'].max()-x['Sales'].min()

sales_data = df.groupby("region").apply(sales_range)
sales_data
```

```python
from numpy import median


df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Salary': [50000,60000,40000,45000,70000,750000],
    'Bonus': [5000,600,2000,2500,8000,85000]
})

result = df.groupby('Department').agg(['mean',"min","max"])
result
```

```python
result = df.groupby('Department').agg({"Salary":["min","max"],
                                       "Bonus":'sum'})
result
```

```python
result = df.groupby('Department').agg(Mean_Sal=("Salary","mean"),
                                      Min_Sal=("Salary","min"),
                                      Max_Sal=("Salary","max"))
result
```

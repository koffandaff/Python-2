sort_values:


 used to sort a DF or series by the values along a specific axis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'name':["alice","bob","charlie","david"],
                   'score':[88,92,85,90]})
df
```

```python
df.sort_values(by="score")
```

```python
df.sort_values(by="score",ascending=False)
```

```python
df.sort_index(ascending=False)
```

apply()

used to run multiple data in a function

```python
s= pd.Series([35,67,90,45])
def pass_fail(x):
    return 'pass' if x>=50 else 'fail'

result = s.apply(pass_fail)
result
```

```python
df =pd.DataFrame({'Physics':[60,70,80],
                  'chemistry':[65,75,85],
                  'maths':[70,80,90]})
result= df.apply(sum)
result
```

```python
result= df.apply(sum,axis=1)
result
```

```python
result= df.apply(lambda x:x.mean())
result
```

```python
df['total']= df.apply(sum,axis=1)
df
```

```python
def grade(total):
    if total>=250:
        return "A"
    elif total>=200:
        return "B"
    else :
        return "C"
    
df["grades"]=df["total"].apply(grade)
df
```

```python
df= pd.DataFrame({'width':[1,2,3],
                  'height':[4,5,6]})
def get_stats(row):
    return [row['width']*row['height'],2*(row['width'])]
df.apply(get_stats, axis=1, result_type='expand')
```

```python
df.apply(get_stats, axis=1, result_type='reduce')
```

```python
df.apply(get_stats, axis=1, result_type='broadcast')
```

```python
def add_values(x,y,z):
    return x+y+z 
df = pd.DataFrame({
    'A': [1,2],
    'B': [10,20]
})
```

```python
df = df.apply(add_values, args=(1,2)) # x will be from dataframe and y,z will be fromt args=
print(df)
```

```python
def add_values(x,m):
    return x*m 
df = df.apply(add_values,m=10) # without args
print(df)
```

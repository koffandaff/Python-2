```python
# 58

def remove_outliers(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")
    return x[(x >= lower_bound) & (x <= upper_bound)]

remove_outliers(df["expenses"])
```

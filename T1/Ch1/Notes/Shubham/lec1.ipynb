## Unit 1 data analysis with pandas and EDA

```python
import pandas as pd
```

```python
a = [1,2,3]
b= pd.Series(a)
print(b)
print(type(b))
print(b[0])
```

```python
# index intialize

a =[1,2,3]
b = pd.Series(a, index=['b','c','d'])
print(b)
print(b['c'])
```

```python
a =(1.3,2,3)
b = pd.Series(a, index=['b','c','d'])
print(b)
print(b['c'])
```

```python
student = ["abc","xyz","mno"]
marks = [90,80,70]
b = pd.Series(marks, index=student,name="marks")
print(b)
```

```python
calories = {'day1': 420, 'day2': 380, 'day3': 390}
cal = pd.Series(calories,name="calories")
print(cal)
```

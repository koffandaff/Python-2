### How to Find Outliers

Interquartile range (IQR) = Q3(75%) - Q1(25%) 

Now using eq we define a range a lower and a hgiher range if the value is out of that range then it's an outlier

lower_value => q1 - 1.5 * IQR 
higer_value => q3 + 1.5 * IQR

```python
import numpy as np
import pandas as pd

def find_outliers(ds,col):
    q1 = ds[col].quantile(0.25) # fidnging 25%
    q3 = ds[col].quantile(0.75)
    IQR = q3 - q1
    low_val = q1 - 1.5 * IQR
    upper_val = q3 + 1.5 * IQR
    print(f'Low: {low_val}, High: {upper_val}')

    ds = ds.loc[(ds[col] < low_val) | (ds[col]> upper_val)]
    return ds

df = pd.read_csv('auto-mpg.csv')
find_outliers(df,'mpg')
```

```python
find_outliers(df,'acceleration')
```

```python
import numpy as np
import pandas as pd

def remove_outliers(ds,col):
    q1 = ds[col].quantile(0.25) # fidnging 25%
    q3 = ds[col].quantile(0.75)
    IQR = q3 - q1
    low_val = q1 - 1.5 * IQR
    upper_val = q3 + 1.5 * IQR
    print(f'Low: {low_val}, High: {upper_val}')

    ds = ds.loc[(ds[col] > low_val) & (ds[col] < upper_val)]
    return ds

df = pd.read_csv('auto-mpg.csv')
remove_outliers(df,'mpg')
```

```python
df = pd.DataFrame({
    'Physics': [60,70,80],
    'Chemistry': [65,75,85],
    'Maths': [70,80,90]
})
result = df.apply(sum) # default Function sum that js adds the values
print(result)
```

```python
aresult = df.apply(sum, axis=1) # add rows data instead of column
aresult
```

```python
# using lambda
meanresult = df.apply(lambda x: x.mean())
print(meanresult)
```

```python
## actual usage

df['total'] = df.apply(sum, axis=1)
df
```

```python
def grade(x):
    if x>=250 :
        return 'A'
    elif x>=200 :
        return 'B'
    else :
        return 'C'

df['grade'] = df['total'].apply(grade)
df
```

```python
# topper
toplist = df.sort_values(by='grade')
toplist
```

```python
import pandas as pd 
df = pd.DataFrame({'width': [1,2],
                    'height': [3,4]})
            
def get_stats(row):
    return [row['width']*row['height'], 2*(row['width']+ row['height'])]
    
```

```python
df.apply(get_stats, axis=1, result_type='expand')
```

```python
df.apply(get_stats, axis=1, result_type='reduce')
```

```python
df.apply(get_stats, axis=1, result_type='broadcast')
```

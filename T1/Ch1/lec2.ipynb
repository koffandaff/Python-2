```python
data = {
    'calories':["111",222,333,444],
    'duration':[50,60,70,80]
}
df = pd.DataFrame(data,index=["day1","day2","day3","day4"])
df
```

```python
data = {'Name':["Alice","Bob","Charlie"],
        'Age':[25,30,35],
        'City':["New York","Los Angeles","Chicago"]}
df = pd.DataFrame(data)
df
```

```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
df.loc[0:1] # last point will be include
```

```python
df.iloc[0:1] # last point will be exclude
```

```python
df[(df.Age>26) & (df.City=="Chicago")]
```

```python
# set_index()

data = {'Name':["Alice","Bob","Charlie"],
        'Age':[25,30,35],
        'City':["New York","Los Angeles","Chicago"]}
df = pd.DataFrame(data)
df
```

```python
df.set_index("Name", drop=False)
```

```python
df.set_index("Name", drop=False,inplace=True)
df
```

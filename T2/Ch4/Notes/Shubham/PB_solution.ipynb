```python
df=pd.read_csv("Location1.csv")
df.isnull().sum()
df.duplicated().sum()
df.columns
```

```python
x=df[['windspeed_10m','windspeed_100m']]
y=df[['Power']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
p=[3,4,5,6]
m_values=[]
for i in p:
    poly=PolynomialFeatures(degree=i)
    x_train_poly=poly.fit_transform(x_train)
    model=LinearRegression()
    model.fit(x_train_poly,y_train)
    y_pred=model.predict(poly.transform(x_test))
    print(y_pred)
    m=mean_squared_error(y_test,y_pred)
    print(m)
    m_values.append(m)
    r=r2_score(y_test,y_pred)
    print(r)

    plt.scatter(y_test,y_pred,label=f"Degree {i}")
plt.xlabel("Actual Power")
plt.ylabel("Predicted Power")
plt.title("Actual vs Predicted Power Values")
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(8,5))
plt.plot(p,m_values,marker='o')
plt.xlabel("Polynomial Degree")
plt.ylabel("Mean Squared Error")
plt.title("MSE vs Polynomial Degree")
plt.grid(True)
plt.show()
```

```python
low=p[np.argmin(m_values)]

print("Lowest mse =",low)
```

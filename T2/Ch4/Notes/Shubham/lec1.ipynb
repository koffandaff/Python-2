```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

# Regression - Model training and Evaluation

- Dependent variable (y) - predict
- Independent variable (x) - used to predict y

simple linear regression: 
* y = mx+c 
* y = m1x1 + m2x2 + m3x3 + c
* y = m.x.^3 + m2x2^2 + m3x3 + c

Evaluation ways
* MAE - mean abosolute error AVG(y-ypred)
* MSE - Mean squared error (AVG(y-ypred)^2)
* R2 score

#### Finding of R2 score
- R2 = 1- (SSres/Sum of square of total)

```python
df = pd.read_csv("./data.csv")
df
```

```python
x = df.iloc[:,0:1] # second xlicing for type change
y = df.iloc[:,-1]
```

```python
lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x)
y_pred
```

```python
# differernce
diff = pd.DataFrame({"Actual": y,"Predication": y_pred})
diff
```

```python
print("Cofficent", lr.coef_)
print("Intercept",lr.intercept_)
```

```python
print("MAE", metrics.mean_absolute_error(y,y_pred))
print("MSE", metrics.mean_squared_error(y,y_pred))
print("R2 Score", metrics.r2_score(y,y_pred))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1, random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
y_pred
```

```python
diff = pd.DataFrame({"Actual": y_test,"Predication": y_pred})
diff
```

```python
print("Cofficent", lr.coef_)
print("Intercept",lr.intercept_)
```

```python
print("MAE", metrics.mean_absolute_error(y_test,y_pred))
print("MSE", metrics.mean_squared_error(y_test,y_pred))
print("R2 Score", metrics.r2_score(y_test,y_pred))
```

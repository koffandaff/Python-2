```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
import math
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

Three category classification


```python
df = pd.read_csv("../../../../PYTHON-2-DATASET-main/AptitudeCommunication.csv")
df
```

```python
x = df[["Aptitude","Communication"]]
y = df["Class"]
```

```python
kn = KNeighborsClassifier(n_neighbors=2)
kn.fit(x,y)
y_pred = kn.predict(x)
y_pred
```

```python
diff = pd.DataFrame({'Actual':y,'Predicted':y_pred})
diff
```

```python
print("accuracy",metrics.accuracy_score(y,y_pred))
```

```python
cm = metrics.confusion_matrix(y,y_pred)
cm
```

```python
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=["S","I","L"])
cm.plot()
plt.show()
```

# Decision Tree

```python
df = pd.read_csv("../../../../PYTHON-2-DATASET-main/DecisionTreeDataset -Num.csv")
df
```

```python
x = df.drop("Job Offered",axis=1)
y = df["Job Offered"]
```

```python
dtree = DecisionTreeClassifier(criterion="log_loss")
model = dtree.fit(x,y)
prediction = model.predict(x)
prediction
```

```python
diff = pd.DataFrame({'Actual':y,'Predicted':prediction})
diff
```

```python
print("Accuracy",metrics.accuracy_score(y,prediction))
```

# 331 - Image Classification using CNN with Transfer Learning

Using a pre-trained MobileNetV2 model on CIFAR-10 dataset

```python
# Step 1: Import necessary libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
# Step 2: Load and preprocess the dataset (CIFAR-10: 10 classes, 32x32 images)
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Class names for CIFAR-10
class_names = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
               'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

# Normalize pixel values to [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Resize images from 32x32 to 96x96 (MobileNetV2 requires minimum 96x96)
x_train_resized = tf.image.resize(x_train, (96, 96))
x_test_resized = tf.image.resize(x_test, (96, 96))

# Convert labels to categorical (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print(f"Training data shape: {x_train_resized.shape}")
print(f"Test data shape: {x_test_resized.shape}")
```

```python
# Step 3: Load pre-trained model (MobileNetV2) WITHOUT top classification layers
base_model = MobileNetV2(
    weights='imagenet',        # Use ImageNet pre-trained weights
    include_top=False,         # Exclude the top classification layer
    input_shape=(96, 96, 3)    # Input shape for our images
)

# Step 4: Freeze the base model layers (so they are not trained again)
base_model.trainable = False
```

```python
# Step 5: Add custom classification layers on top of the base model
x = base_model.output
x = GlobalAveragePooling2D()(x)      # Global average pooling
x = Dense(128, activation='relu')(x)  # Fully connected layer
x = Dropout(0.5)(x)                   # Dropout for regularization
predictions = Dense(10, activation='softmax')(x)  # Output layer (10 classes)

# Step 6: Create the final model
model = Model(inputs=base_model.input, outputs=predictions)
```

```python
# Step 7: Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()
```

```python
# Step 8: Train the model
history = model.fit(
    x_train_resized, y_train_cat,
    epochs=5,
    batch_size=32,
    verbose=1,
    validation_data=(x_test_resized, y_test_cat)
)
```

```python
# Step 9: Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(x_test_resized, y_test_cat, verbose=0)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
# Step 10: Predict the class of a new image
sample_image = x_test_resized[0:1]  # Take the first test image
prediction = model.predict(sample_image)
predicted_class = np.argmax(prediction)
actual_class = np.argmax(y_test_cat[0])

print(f"Predicted Class: {class_names[predicted_class]}")
print(f"Actual Class:    {class_names[actual_class]}")
```

```python
# Step 11: Plot training history
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Accuracy plot
axes[0].plot(history.history['accuracy'], label='Train Accuracy')
axes[0].plot(history.history['val_accuracy'], label='Validation Accuracy')
axes[0].set_title('Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()

# Loss plot
axes[1].plot(history.history['loss'], label='Train Loss')
axes[1].plot(history.history['val_loss'], label='Validation Loss')
axes[1].set_title('Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()

plt.tight_layout()
plt.show()
```
